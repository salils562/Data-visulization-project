# Topic: Heart prediction failure dataset 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#reading the dataset
heart_df=pd.read_csv('D:\\Python csv files\\heart.csv')
pd.set_option('display.max_rows',None)
heart_df
```

# cleaning the dataset

```python
heart_df.interpolate(how='index')
```

# Peoples more than age 40-years having heart disease

```python
heart_df[heart_df['Age']>40].sort_values(by=['Age'],ascending=True)
```

## Male having heart disease 

```python
heart_df[heart_df['Sex']=='M'].sort_values(by=['Age'],ascending=True)
```

# Female having heart dieseases

```python
heart_df[heart_df['Sex']=='F'].sort_values(by=['Age'])
```

## Graph of heart dieseases 

```python
plt.figure(figsize=(8,8))
sns.barplot(x='Sex',y='Age',data=heart_df,linewidth=8)
plt.show()
```

# cholestrol graph 

```python
plt.figure(figsize=(7,7))
sns.barplot(x='Sex',y='Cholesterol',data=heart_df,edgecolor='m',linewidth=4)
plt.title('Cholestrol for male and female',fontsize=19)
plt.xlabel('Male and Female',color='c',fontsize=14)
plt.ylabel('Cholestrol data',fontsize=14,color='c')
plt.show()
```

# Scatter Graph between Cholestrol and Chestpain type

```python
plt.figure(figsize=(9,7))
sns.scatterplot(x='ChestPainType',y='Cholesterol',data=heart_df,hue='Sex',size='Sex',sizes=(100,200),marker='o')
plt.title('Cholestrol-chestpain graph between male and female',fontsize=20,color='m')
plt.show()
```

# Peoples having resting ECG normal

```python
heart_df[heart_df['RestingECG']=='Normal']
```

# Total peoples suffering from heart diesease

```python
tot=heart_df[heart_df['HeartDisease']==0].value_counts().sum()
totf=heart_df[heart_df['HeartDisease']==1].value_counts().sum()
print('Number of peoples not suffering from heartdiesease are:',tot)
print('Number of peoples suffering from heartdiesease are:',totf)
```

# Graph for males and females for  rising heart-rate

```python
plt.figure(figsize=(8,5))
sns.lineplot(x='Sex',y='MaxHR',data=heart_df,color='g',palette='hot',label='Rising-rate')
plt.title('Heart-rate',fontsize=14,color='m')
plt.xlabel('Sex',fontsize=14,color='r')
sns.set()
plt.ylabel('Max-heartrate',fontsize=14,color='r')
plt.show()
```

# Pie graph between peoples with heart-dieseases

```python
plt.figure(figsize=(6,7))
heart_patient=heart_df[heart_df['HeartDisease']==0].value_counts().sum()
nonhea_patient=heart_df[heart_df['HeartDisease']==1].value_counts().sum()
persons=['Heart-patient','Non-heart patient']
p=[heart_patient,nonhea_patient]
text={'fontsize':15}
plt.pie(p,labels=persons,explode=[0.03,0.03],autopct='%0.1f%%',textprops=text)
plt.show()
```
